### Data Cleaning - Missing Value

```python
df.head()
```

```python
df.isnull().sum()
```

```python
def proc(x):
    if x=='-' or not x:
        return None 
    elif len(x.split('-')) ==2:
        return ((float(x.split('-')[0])) + (float(x.split('-')[1]))) / 2
    return float(x.split('-')[0])
    
```

```python
df_raw['overall'].unique()
```

```python
for x in ['overall', 'resources', 'engagement', 'outcomes', 'enviroment']:
        df[x] = df[x].apply(proc)
        
        
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.tail()
```

```python
df.dropna(axis=0, thresh=6, inplace=True) # drop the row, at least 6 non missing value

```

```python
df.shape
```

```python
# when resource is NULL, overal socre distribute on 31.95, 42.7, 46,.6

df[df['resources'].isna()]['overall'].value_counts() / df.shape[0]
```

```python
df.tail()
```

```python
# impute value by 'overall' group

df[['resources', 'engagement', 'outcomes', 'enviroment']] = \
df.groupby('overall').transform(lambda x: x.fillna(x.mean()))
```

```python
df.tail()
```

### Data Preprocessing - outlier

```python
# 5 IQR rules
q1 = np.quantile(df['overall'], 0.25)
q3 = np.quantile(df['overall'], 0.75)
# finding the iqr region
iqr = q3-q1
# finding upper and lower whiskers
upper_bound = q3+(1.5*iqr)
lower_bound = q1-(1.5*iqr)
print(iqr, upper_bound, lower_bound)
```

```python
df.loc[(df['overall']<=lower_bound)|(df['overall']>=upper_bound)] 
```

### Data Preprocessing - Imbalance Data

```python
df['label'] = np.where(\
    df['rank'].isin([str(x) for x in range(0,51)])\
     ,1 , 0 )
```

```python
df.tail()
```

```python
df['label'].value_counts()
```

```python
### SMOTE

from imblearn.over_sampling import SMOTE 
```

```python
X_train = df[['overall', 'resources', 'engagement', 'outcomes','enviroment']]
y_train = df[['label']] # class/target column

```

```python
sm = SMOTE()
X_train_oversample , y_train_oversample = \
                            sm.fit_resample(X_train, y_train)


pd.DataFrame(X_train_oversample, columns=X_train.columns)
```
